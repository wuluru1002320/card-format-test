```python
def combine_with_description(type, typeDescription):
    """
    結合employeeId和隨機的employeeIdDescription，如果有的話，使用隨機的分隔符連接它們。

    :param employeeId: 員工編號的陣列
    :param employeeIdDescription: 員工編號描述的陣列
    :return: 組合後的字串陣列
    """
    combined_strings = []

    for i in type:
        combined_string = i
        if typeDescription and random.choice([True, False]):
                # 隨機選擇一個描述
                description = random.choice(typeDescription);
                # 使用隨機的分隔符連接
                separator = random.choice([' ', ' - '])
                if (separator == ' '):
                     combined_string += separator + '(' +description + ')'
                else:
                    combined_string += separator + description
        combined_strings.append(combined_string)
    # 將新生成的字串與傳入的type陣列結合成新的陣列
    combined_with_type = combined_strings + type
    unique_set = set(combined_with_type);
    unique_list = list(unique_set)
    return unique_list

# 呼叫函數來生成組合的字串陣列
combined_employeeId = combine_with_description(employeeId, employeeIdDescription)
combined_name = combine_with_description(name, nameDescription)
combined_id = combine_with_description(idCard, idCardDescription)
```

```python
def combine_and_shuffle_arrays(data_dict):
    """
    將多個陣列合併並隨機打亂，同時生成對應的label陣列。

    :param data_dict: 陣列字典，其中鍵是類別名稱，值是陣列
    :return: train陣列和label陣列
    """
    combined_data = []
    label = []

    for category, values in data_dict.items():
        for value in values:
            combined_data.append(value)
            label.append(type_dict[category])

    # 使用random.shuffle()函數隨機打亂combined_data和label
    combined_and_label = list(zip(combined_data, label))
    random.shuffle(combined_and_label)

    # 將結果拆分回train和label陣列
    train, label = zip(*combined_and_label)

    return list(train), list(label)

type_dict = {
    'employeeId':0,
    'name':1,
    'gender':2,
    'idCard':3,
    'company':4,
    'department':5,
    'email':6,
    'workTime':7,
    'others':8
}

# 定義陣列和對應的類別編號
data_dict = {
    'employeeId':combined_employeeId,
    'name':combined_name,
    'gender':gender,
    'idCard':combined_id,
    'company':company,
    'department':department,
    'email':email,
    'workTime':workTime,
    'others':others,
}

# 呼叫函數來生成train和label
data, labels = combine_and_shuffle_arrays(data_dict)

# for string in train_data:
#     print(string)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# 選擇預訓練的BERT模型和分詞器
model_name = "bert-base-uncased"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)

# 分詞、向量化、並處理每個輸入字串
input_ids = []
attention_masks = []

# 計算所有文本中的最大長度
max_length = max(len(tokenizer.tokenize(t)) for t in data)
# 確保所有文本的長度相同，使用填充
for text in data:
    tokens = tokenizer(text, padding="max_length", truncation=True, return_tensors="pt")
    input_ids.append(tokens.input_ids)
    attention_masks.append(tokens.attention_mask)

# 將 input_ids 和 attention_masks 轉換為 PyTorch 張量
input_ids = torch.cat(input_ids, dim=0)
attention_masks = torch.cat(attention_masks, dim=0)

# 通過模型獲取向量
with torch.no_grad():
    outputs = model(input_ids, attention_mask=attention_masks)
    embeddings = outputs.last_hidden_state
```

```python
import tensorflow as tf
import numpy as np

# 提取[CLS]标记向量，假设BERT输出的第一个标记是[CLS]
cls_vector = embeddings[:, 0, :]  # 提取每个样本的[CLS]向量

# 創建一個全連接神經網絡模型
model = tf.keras.Sequential([
    # 不再使用Flatten层，直接使用[CLS]向量作为输入
    tf.keras.layers.Input(shape=(768,)),  # 假设輸入的BERT向量大小為768
    
    # 添加一個全連接隱藏層
    tf.keras.layers.Dense(128, activation='relu'),  # 128是隱藏層的神經元數量，可以根據需要調整
    
    # 添加輸出層，通常使用softmax激活函數進行多類別分類
    tf.keras.layers.Dense(9, activation='softmax')  # num_classes是分類任務的類別數
])

# 編譯模型，定義損失函數和優化器
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # 適用於多類別分類任務
              metrics=['accuracy'])

# 查看模型結構
# model.summary()

# 将 PyTorch 张量转换为 NumPy 数组
train_data = cls_vector.cpu().numpy()

# 将labels转换为NumPy数组
train_labels = np.array(labels)

# # 訓練模型
history = model.fit(train_data, train_labels, 
                    epochs=50,
                    batch_size=12,
                    validation_split=0.2
                    )

# # 在測試集上評估模型性能
test_loss, test_accuracy = model.evaluate(train_data, train_labels)
print(f'測試準確度：{test_accuracy}')
```
